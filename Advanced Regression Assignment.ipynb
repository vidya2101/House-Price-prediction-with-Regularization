### Subtask 2.1: `Dropping or Imputing` data (Handling `NaN/Null` Values)
Removing or imputing data is performed in steps. While pre-processing, we shall follow the below thumb rules:
1. __`Percentage missing values in columns:`__
    - If `missing values > 40%`, drop columns directly
    - If `< 40%`, Inpute if feasible otherwise drop
      
      
2. __`Percentage missing values in rows:`__
    - Delete rows with all null values
    - Delete all rows with `> 5` null values
    
    
3. `Re-inspect` the data 

4. Repeat the process until we have `all clean (no null value) data`

__*Note:*__ *`Linear regression of any form` is a numerical technique and it cannot deal with `NaN/Null` Values*

#### `Imputing using probabilities` 
1. We know that, if there is a large number of data. The probability of a particular outcome is given by `no of times that particular outcome came/ total number of outcomes.`
2. In this case, we have sufficient number of data points.
3. For time being, to impute data as randomly as possible. We assume that, `outcome of each column is independent of other columns.`
4. After making that assumption, `percent count or Frequency` will be the probability of that respective `category` as an outcome.
5. So, let us distribute the `NaN data` using the same `multinomial probabilities`.
6. This helps us to have the existing data, `same percentage of each category even after imputation`.

Let us do it on the previously discussed columns. `GarageType`, `GarageFinish`, `BsmtQual` and `BsmtFinType1`

```python
# Step 1. Collect all the columns to impute by this method as a list
columns_impute = ['GarageType','GarageFinish', 'BsmtQual', 'BsmtFinType1']

# Step 2: Compute the total number of null values in the listed columns
total_null_values = df_house[columns_impute].isnull().sum()
total_null_values
```

```python
# Step 3: Probabilities (Occurance) of each category in each column
Probabilities = {}
list_categories = {}
rounded_probabilities = {}
for i in columns_impute:
    Probabilities[i] = df_house[i].value_counts()/(df_house.shape[0] - total_null_values[i])
    list_categories[i] = list(df_house[i].value_counts().index)
    rounded_probabilities[i] = list(np.around(Probabilities[i].values, decimals = 2))
rounded_probabilities
```

```python
list_categories
```

```python
# Step 3: List of how many values to be replaced by which category
null_frequency = {}
for i in columns_impute:
    null_frequency[i] = list(np.random.multinomial(total_null_values[i], rounded_probabilities[i]))
null_frequency
```

##### Step 4: Generatig categorical data with their respective probabilities for each column

```python
#Step 4: Generating Categorical Data of Tags using the distribution above
# a: For GarageType 
list_categories_fill = list(list_categories['GarageType'])
list_GarageType = []
for i in range(0,len(list_categories_fill)):
     
    for j in range(0,null_frequency['GarageType'][i]):
        list_GarageType.append(list_categories_fill[i])
```

```python
# b: For GarageFinish 
list_categories_fill = list(list_categories['GarageFinish'])
list_GarageFinish = []
for i in range(0,len(list_categories_fill)):
     
    for j in range(0,null_frequency['GarageFinish'][i]):
        list_GarageFinish.append(list_categories_fill[i])
```

```python
#c: For BsmtQual 
list_categories_fill = list(list_categories['BsmtQual'])
list_BsmtQual = []
for i in range(0,len(list_categories_fill)):
     
    for j in range(0,null_frequency['BsmtQual'][i]):
        list_BsmtQual.append(list_categories_fill[i])
```

```python
#d: For BsmtFinType1 
list_categories_fill = list(list_categories['BsmtFinType1'])
list_BsmtFinType1 = []
for i in range(0,len(list_categories_fill)):
     
    for j in range(0,null_frequency['BsmtFinType1'][i]):
        list_BsmtFinType1.append(list_categories_fill[i])
```

##### Step 5: Adding imputed data to the dataframe

```python
# Step 5:
# For GarageType
#Make a series out of "list_tags" so that we can randomize the values
series_GarageType = pd.Series(list_GarageType)
series_GarageType = series_GarageType.sample(frac=1).reset_index(drop=True)
#series_tags.values

# Step: nan list
index1 = list(df_house[df_house['GarageType'].isnull()].index)

# Imputing Tag values wit the generated series
df_house['GarageType'].loc[index1] = list(series_GarageType.values)
#df_leads.head(20)
```

### Subtask 4.3: `Re-Scaling` Data:

####  `Train-Test`  Split  
It is advised to re-scale data `after the train-test split`. <br> 
__`Reason:`__ While re-scaling data, we will use values like `minimun/maximum/variance/mean`. If we use these values before test-train split. We are including the influence of test data also in scaling which evidently makes test data a seen data.
- `Example:` Suppose say there is a variable $x_1$,
    - Train data range: (-0.8, 0.8)
    - Test data range: (-1, 1)
    -  Before split, range: (-1,1)
<br> <br>
- Now  if we do `min-max scaling` before __train test split__. <br> <br>
    - Formula: $x_1^{scaled} = \frac{x_1^{value} - x_1^{min}}{x_1^{max} - x_1^{min}}$
    - Train data will be split using (-1,1) while these values do not exist in train data after splitting.
    - Also, we are scaling according to the test data
<br> <br>
- This is not advisable. So `train-test split` is always performed before scaling to avoid these effects.
- Re-Scaling is performed __`only on quantitative variables`__. 

I am taking `train-test split in 0.7:0.3 ratio` since data is less in number.
